# 特征提取

```python
from sklearn.feature_extraction import DictVectorizer
```

## 一、字典特征提取

```python
data = [{'city': '北京','temperature':100}, {'city': '上海','temperature':60}, {'city': '深圳','temperature':30}]
```

### 1. sparse=True，返回 sparse 矩阵，节省内存

```python
# 1、实例化一个转换器类
transfer = DictVectorizer(sparse=True)
# 2、调用fit_transform
print(transfer.fit_transform(data))
```

### 2. sparse=False，

```python
transfer = DictVectorizer(sparse=False)
print(transfer.fit_transform(data))
```

### +  打印特征名字

```python
print(transfer.get_feature_names())
```

> 上面是独热编码

# 二、文本特征提取

```python
data = ["life is short,i like like python", "life is too long,i dislike python"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
transfer = CountVectorizer()
```

### 1. 返回 sparse 矩阵，节省内存

```python
print(transfer.fit_transform(data))
```

### 2. 返回 真实矩阵，利用toarray()进行sparse矩阵转换array数组

```python
print(transfer.fit_transform(data).toarray())
```

### + 返回特征名字

```python
print("返回特征名字：\n", transfer.get_feature_names())
```

### + 过滤单词 如：'is'

```python
transfer = CountVectorizer(stop_words=['is'])
transfer.fit_transform(data).toarray()
print("返回特征名字：\n", transfer.get_feature_names())
```

# 三、中文文本特征提取

```python
import jieba
```

### + 简单示例， 用结巴对中文字符串进行分词

```python
text = '我爱北京天安门'
print(' '.join(list(jieba.cut(text))))
```

### + 特征提取

```python
texts = ["一种还是一种今天很残酷，明天更残酷，后天很美好，但绝对大部分是死在明天晚上，所以每个人不要放弃今天。",
            "我们看到的从很远星系来的光是在几百万年之前发出的，这样当我们看到宇宙时，我们是在看它的过去。",
            "如果只用一种方式了解某样事物，你就不会真正了解它。了解事物真正含义的秘密取决于如何将其与我们所了解的事物相联系。"]

# 创建分词匿名函数
cut_word = lambda text:' '.join(list(jieba.cut(text)))

# 对原始数据进行jieba分词
new_texts = [cut_word(text) for text in texts]

# 创建分类器
transfer = CountVectorizer()

# 返回频次数组
print(transfer.fit_transform(new_texts).toarray())

# 返回特征词
print(transfer.get_feature_names())
```

# 四、Tf-idf文本特征提取

```python
from sklearn.feature_extraction.text import TfidfVectorizer

texts = ["一种还是一种今天很残酷，明天更残酷，后天很美好，但绝对大部分是死在明天晚上，所以每个人不要放弃今天。",
            "我们看到的从很远星系来的光是在几百万年之前发出的，这样当我们看到宇宙时，我们是在看它的过去。",
            "如果只用一种方式了解某样事物，你就不会真正了解它。了解事物真正含义的秘密取决于如何将其与我们所了解的事物相联系。"]

# 创建分词匿名函数
cut_word = lambda text:' '.join(list(jieba.cut(text)))

# 对原始数据进行jieba分词
new_texts = [cut_word(text) for text in texts]

# 创建分类器
transfer = TfidfVectorizer(stop_words=['一种', '不会', '不要'])

# 返回频次数组
print(transfer.fit_transform(new_texts).toarray())

# 返回特征词
print(transfer.get_feature_names())
```

## 1. 降维 - 特征选择 - 过滤式 - 方差选择法

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
data = pd.read_csv('./data/factor_returns.csv')

print(data[data.columns[1:-2]].shape)

# 1、实例化一个转换器类
transfer = VarianceThreshold(threshold=1)

# 2、调用fit_transform
new_data = transfer.fit_transform(data[data.columns[1:-2]])

# 3、删除低方差特征的结果
print(new_data.shape)
```

## 2. 降维 - 特征选择 - 过滤式 - 相关系数

```python
from scipy.stats import pearsonr

data = pd.read_csv('./data/factor_returns.csv')
```

```python
factor = ['pe_ratio', 'pb_ratio', 'market_cap', 'return_on_asset_net_profit', 'du_return_on_equity', 'ev',
              'earnings_per_share', 'revenue', 'total_expense']

new_data = [(factor[i], factor[j + 1], pearsonr(data[factor[i]], data[factor[j + 1]])[0]) for i in range(len(factor)) for j in range(i, len(factor) - 1)]

for dt in new_data:
    print("指标 {} 与指标 {} 之间的相关性大小为 {} ".format(*dt))
```

#### 画图来观察结果

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 8), dpi=100)
plt.scatter(data['revenue'], data['total_expense'])
plt.show()
```

```python
# 绘制矩阵图
pd.scatter_matrix(data[data.columns[1:-2]],figsize=(40,24))
plt.show()
```

## 3. 降维 - 主成分分析（PCA）

```python
from sklearn.decomposition import PCA
```

```python
data = [[2,8,4,5], [6,3,0,8], [5,4,9,1]]
```

```python
 # 1、实例化PCA, 小数—保留百分之多少信息
transfer = PCA(n_components=0.9)
# 2、调用fit_transform
data1 = transfer.fit_transform(data)

print("保留90%的信息，降维结果为：\n", data1)
```

```python
# 1、实例化PCA, 整数——指定降维到的维数
transfer2 = PCA(n_components=3)
# 2、调用fit_transform
data2 = transfer2.fit_transform(data)
print("降维到3维的结果：\n", data2)
```
