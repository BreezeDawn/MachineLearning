# 十四、缺失值处理

## 1.判断数据是否为NaN

````python
pd.isnull(df), pd.notnull(df)
````

## 2.处理方式

+ ## 删除存在缺失值的：dropna(axis='rows')
### 使用dropna的前提是，缺失值的类型必须是np.nan
### 注：不会修改原数据，需要接受返回值

+ ## 替换缺失值：fillna(value, inplace=True)
### value：替换成的值
### inplace：
        True：会修改原数据
        False：不替换修改原数据，生成新的对象

+ ## 不是缺失值nan，有默认标记的


## 3. 案例

````python
movie = pd.read_csv("./IMDB-Movie-Data.csv")
````

````python
pd.notnull(movie)
````

````python
# 把一些其它值标记的缺失值，替换成np.nan
movie = movie.replace(to_replace='?', value=np.nan)
````

````python
movie_new = movie.dropna()  # 删除缺失值
pd.notnull(movie_new)
````

````python
# 替换存在缺失值的样本的两列
# 替换填充平均值，中位数
movie['Revenue (Millions)'].fillna(movie['Revenue (Millions)'].mean(), inplace=True)
movie['Metascore'].fillna(movie['Metascore'].mean(), inplace=True)
pd.notnull(movie)
````

# 十五、数据离散化
## 为什么要离散化
### 连续属性离散化的目的是为了简化数据结构，数据离散化技术可以用来减少给定连续属性值的个数

+ ## 案例：股票的涨跌幅离散化

````python
data = pd.read_csv("./stock_day.csv")
p_change= data['p_change']
````

````python
data
````

````python
# 自行分组
qcut = pd.qcut(p_change, 10)
# 计算分到每个组数据个数
qcut.value_counts()
````

````python
# 自己指定分组区间
bins = [-100, -7, -5, -3, 0, 3, 5, 7, 100]
p_counts = pd.cut(p_change, bins)
p_counts.value_counts()
````

````python
dummies = pd.get_dummies(p_counts, prefix="涨跌幅")
dummies.head()
````

## 2.pd.merge实现合并

````python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                        'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                        'key2': ['K0', 'K0', 'K0', 'K0'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']})
````

````python
left
````

````python
right
````

````python
# 默认内连接，只连接共有的部分
result = pd.merge(left, right, on=['key1', 'key2'])
result
````

````python
# 保证 left 表中的数据齐全
result = pd.merge(left, right, how='left', on=['key1', 'key2'])
result
````

````python
# 保证 left 表中的数据齐全
result = pd.merge(left, right, how='right', on=['key1', 'key2'])
result
````

````python
# 全连接
result = pd.merge(left, right, how='outer', on=['key1', 'key2'])
result
````

# 十七、交叉表与透视表
### 研究两个分类变量间的关系

 + ## 使用crosstab(交叉表)实现

````python
data
````

````python
date = pd.to_datetime(data.index).weekday
data['week'] = date
data['week'].head()
````

````python
# 2、把p_change按照大小分类，以0为界限
data['posi_neg'] = np.where(data['p_change'] > 0, 1, 0)
````

````python
# 通过交叉表找寻两列数据的关系
count = pd.crosstab(data['week'], data['posi_neg'])
count
````

````python
count.div(count.sum(axis=1), axis=0)
````

````python
count.div(count.sum(axis=1), axis=0).plot(kind='bar',stacked=True,figsize=(10,4))
plt.show()
````

+ ##  使用pivot_table(透视表)实现

````python
# 通过透视表，将整个过程变成更简单一些
data.pivot_table(['posi_neg'], index=['week'])
````
